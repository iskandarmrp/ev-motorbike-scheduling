```python
import random
import pickle

# 10 battery swap station, masing-masing dengan 2 slot
battery_swap_station_demo = [
    [
        [random.randint(40, 100), random.randint(11, 20)],
        [random.randint(40, 100), random.randint(31, 40)]
    ]
    for _ in range(10)
]

# 1000 EV, setiap EV memiliki energy_distance dan travel_time untuk 10 station
ev_demo = {}
for i in range(1000):
    battery_now = random.randint(20, 100)
    battery_cycle = random.randint(20, 100)
    energy_distance = [random.randint(10, 40) for _ in range(10)]
    travel_time = [random.randint(10, 40) for _ in range(10)]

    # Setiap kelipatan 50, buat EV memiliki jadwal tetap (fixed schedule)
    if i % 50 == 0:
        station = random.randint(0, 9)  # hanya sampai station index 9
        slot = random.randint(0, 1)
        ed = energy_distance[station]
        tt = travel_time[station]
        exchanged_battery = max(0, battery_now - ed)
        received_battery = random.randint(80, 100)
        received_battery_cycle = random.randint(1, 100)
        waiting_time = random.randint(5, 15)

        swap_schedule = {
            "assigned": True,
            "battery_now": battery_now,
            "battery_cycle": battery_cycle,
            "battery_station": station,
            "slot": slot,
            "energy_distance": ed,
            "travel_time": tt,
            "waiting_time": waiting_time,
            "exchanged_battery": exchanged_battery,
            "received_battery": received_battery,
            "exchanged_battery_cycle": battery_cycle,
            "received_battery_cycle": received_battery_cycle,
        }
    else:
        swap_schedule = {}

    ev_demo[i] = {
        "battery_now": battery_now,
        "battery_cycle": battery_cycle,
        "energy_distance": energy_distance,
        "travel_time": travel_time,
        "swap_schedule": swap_schedule,
    }

with open("ev_battery_swap_random_dataset.pkl", "wb") as f:
    pickle.dump((battery_swap_station_demo, ev_demo), f)
```
